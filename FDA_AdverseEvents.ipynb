```python
import numpy as np
import pandas as pd
import requests
import json
from pprint import pprint
```

```python
url = "https://api.fda.gov/device/event.json?search="
date="date_received:[2020101+TO+20201231]+"
#date="suppl_dates_mfr_received:[01/01/2020+TO+12/31/2020]+"
#mfr="device.manufacturer_d_name:*medtronic*+OR+*boston*+"
name="device.generic_name:*pacemaker&"
limit="limit=1000"

response = requests.get(url + date + name +  limit)
data1 = response.json()
pprint(data1)
```

```python
results=data1['results']
```

```python
#count the number of records

print(len(data1['results']))
```

```python
#pull out the date of events and create a list named "dates"
dates=[]
for item in range(len(data1['results'])):
    print(data1['results'][item]['date_received'])
    dates.append(data1['results'][item]['date_received'])
    #dates.sort
   # print(dates)
```

```python
#Create a list of the manufacturers
manufacturers=[]
for row in range(len(data1['results'])):
    print(results[row]['device'][0]['manufacturer_d_name'])
    manufacturers.append(results[row]['device'][0]['manufacturer_d_name'])
       
```

```python
age=[]
for row in range(len(data1['results'])):
    #print(results[row]['patient'][0]['patient_age'])
    age.append(results[row]['patient'][0]['patient_age'])
```

```python
#pull out the date of events and create a list named "dates"
eventType=[]
for event in range(len(data1['results'])):
    print(data1['results'][event]['event_type'])
    eventType.append(data1['results'][event]['event_type'])
```

```python
Report=[]
for report in range(len(data1['results'])):
    print(data1['results'][report]['report_number'])
    Report.append(data1['results'][report]['report_number'])
```

```python
#Create a dataframe from the variables created for Date, Manufacturer, Event Type

df=pd.DataFrame(
    {
        'Dates':dates,
        'Manufacturer':manufacturers,
        'Event Type': eventType,
        'Age': age,
        'Report #': Report
        
        
})
df.head(20)
```
